```python
plt.magnitude_spectrum(yfft) 
plt.show() 
```

Observalos o efeito de spectral leakage, onde o sinal com vazamento tem menor amplitude e uma resposta de frequência mais ampla (t par) do que um sinal sem vazamento (t impar). Isso dificulta quantificar o sinal corretamente no domínio de frequência. O vazamento é criado por uma transformação digital do Fourier de dados não periódicos e causa uma redistribuição do sinal sobre toda a faixa de frequência da medição um janelamento pode ajudar a minimizar os efeitos do vazamento.

c. Repita o item b somente utilizando o sinal de 1s. utilizando as seguintes
Janelas triangular, Hanning, hamming, e flattop. Discuta as diferenças
entre as janelas no espectro de magnitude.

```python
y = s1
f = 16
k = 0
tf = 1

plt.figure(figsize=(20,20))
plt.subplot(421)

win = np.ones_like(y)
print(win.shape[0])
win[int(win.shape[0]/2)::] = 0

plt.plot(np.arange(y.shape[0]),y*win)
plt.plot(np.arange(y.shape[0]),win)
plt.title('sinal + janela do tipo caixa')

plt.subplot(422)
yfft = np.fft.fft(y*win,(k+1)*y.shape[0])/y.shape[0]
yfft = np.abs(yfft)
yfft = np.fft.fftshift(yfft)
freq = np.arange(0,fs,fs/yfft.shape[0])-fs/2
plt.plot(freq,yfft)
plt.xlim([10,20])

# janela triangular 
t = np.arange(0,tf,1/fs)
y = np.sin(2*np.pi*f*t)
plt.subplot(423)

win = np.bartlett(y.shape[0])

plt.plot(np.arange(y.shape[0]),y*win)
plt.plot(np.arange(y.shape[0]),win)
plt.title('sinal + janela do tipo triangular')


plt.subplot(424)
yfft = np.fft.fft(y*win,(k+1)*y.shape[0])/y.shape[0]
yfft = np.abs(yfft)
yfft = np.fft.fftshift(yfft)
freq = np.arange(0,fs,fs/yfft.shape[0])-fs/2
plt.plot(freq,yfft)
plt.xlim([10,20])

# janela hanning
t = np.arange(0,tf,1/fs)
y = np.sin(2*np.pi*f*t)
plt.subplot(425)

win = np.hanning(y.shape[0])

plt.plot(np.arange(y.shape[0]),y*win)
plt.plot(np.arange(y.shape[0]),win)
plt.title('sinal + janela do tipo hanning')

plt.subplot(426)
yfft = np.fft.fft(y*win,(k+1)*y.shape[0])/y.shape[0]
yfft = np.abs(yfft)
yfft = np.fft.fftshift(yfft)
freq = np.arange(0,fs,fs/yfft.shape[0])-fs/2
plt.plot(freq,yfft)
plt.xlim([10,20])

# janela hamming
t = np.arange(0,tf,1/fs)
y = np.sin(2*np.pi*f*t)
plt.subplot(427)

win = np.hamming(y.shape[0])

plt.plot(np.arange(y.shape[0]),y*win)
plt.plot(np.arange(y.shape[0]),win)
plt.title('sinal + janela do tipo hamming')


plt.subplot(428)
yfft = np.fft.fft(y*win,(k+1)*y.shape[0])/y.shape[0]
yfft = np.abs(yfft)
yfft = np.fft.fftshift(yfft)
freq = np.arange(0,fs,fs/yfft.shape[0])-fs/2
plt.plot(freq,yfft)
plt.xlim([10,20])
```

Podemos observar que a diferença entre as janelas é justamente na precisão do resultado no espectro. Sendo, a janela hanning ligeiramente melhor em relação as outras exemplificadas.

B) Defina a frequência de amostragem mínima para a preservação do áudio.
Justifique a sua resposta.

*A taxa de amostragem deve ser pelo menos duas vezes maior que a frequência que se deseja registrar. Esse é o Teorema de Nyquist ou o Teorema da Amostragem, extremamente importante para estabelece o critério adequado para a amostragem do sinal desejado e que seja possível recuperar o sinal original. Nyquist provou que, o espectro de um sinal não será afetado pela sua discretização se esta ocorrer com uma frequência de amostragem (fs) duas vezes maior que a maior frequência do sinal.*

*fs >= 2fm*

C) Realize o processo de downsampling no sinal de forma a preservar o espectro
de frequência do sinal. Justifique a escolha do parâmetro n no downsampling.

```python
plt.figure(figsize=(20,10))
plt.plot(freq,np.absolute(yfft))
plt.xlim([0,fs/2])
```

```python
# "n" é um valor inteiro e define a quantidade de amostras (exemplo, se n=2 o downsampling tera metade das amostras do sinal original), 
# o "n" precisa ser escolhido de tal forma que o sinal não sofra aliasing, ou seja, que fique preservado o gráfico da imagem anterior. Sendo assim:
# fs = 44.5kHz, divido por 4 temos 11.125kHz, mais que 5 teremos aliasing. 

n = int(np.floor(fs/8000)) #8000 escolhido pela análise no gráfico anterior, que é onde aparentemente já não tem mais sinal, apenas ruído.
ydownsampling = y[::n]
tdownsampling = t[::n]
fsd = fs/n

# janela hamming
k = 0
t = tdownsampling
y = ydownsampling

win = np.hamming(y.shape[0])

yfft = np.fft.fft(y*win,(k+1)*y.shape[0])/y.shape[0]
yfft = np.abs(yfft)
yfft = np.fft.fftshift(yfft)
freq = np.arange(0,fsd,fsd/yfft.shape[0])-fsd/2

plt.figure(figsize=(20,10))
plt.plot(freq,np.absolute(yfft))
plt.xlim([0,fs/2])
```
