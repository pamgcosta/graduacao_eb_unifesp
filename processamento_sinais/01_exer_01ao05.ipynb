3) Escreva um programa que compare duas listas ( lista-a e lista-b) e
imprima:

a. Os valores comuns as duas listas

```python
#Exemplo de listas
lista_a = [4, 2, 3, 1]
lista_b = [3, 4, 5, 6]
#Lista em comum
lista_k = []

for i in lista_a:
  for j in lista_b:
    if i == j:
      lista_k.append(i)

print(lista_k)
```

b. Os valores que só existem na lista-a

```python
#Lista dos itens exclusivos em a
lista_aex = []

for i in lista_a:
  if i not in lista_k:
    lista_aex.append(i)

print(lista_aex)
```

c. Os valores que só existem na lista-b

```python
#Lista dos itens exclusivos em b
lista_bex = []

for i in lista_b:
  #for j in lista_k:
  if i not in lista_k:
    lista_bex.append(i)

print(lista_bex)
```

d. Uma lista com os elementos não repetidos.

```python
#Lista dos itens exclusivos entre as listas a e b
lista_ex = lista_bex + lista_aex

print(lista_ex)
```

4) Crie uma rotina na qual calcula o valor do Euler e o seno
hiperbólico a partir da série de Taylor ( 20 primeiros termos). OBS:
essa função não pode conter funções de loop tal como: for while.
Dica: para calcular o fatorial de um número utilize a biblioteca
scipy.special.

```python
import scipy.special
import numpy as np
```

Para o cálculo de exponencial (sem funções de loop e utilizando scipy.special)



```python
#Valor para x
x = 2
#Vetor com a posição dos 20 primeiros termos "n" (0-19)
n = np.arange(20)
#Vetor com o valor de x nas 20 posições de cálculo
x =  np.ones(20, dtype=int)*x
#Valor de x^n (denominador) em seus 20 termos
y = x**n
#Fatorial de "n" em seus 20 termos (numerador)
fatorial = scipy.special.factorial(n)
#Expressão final
exp = y/fatorial
exp = sum(exp) 
#Resultado
print("O valor do exponencial é:", exp)
```

Para o cálculo do seno hiperbólico (sem funções de loop e utilizando scipy.special)

```python
#Valor para x
x = 2
#Vetor com a posição dos 20 primeiros termos "n" (0-19)
n = np.arange(20)
#Selecionando apenas os valores impares
n = n[1::2]
#Vetor com o valor de x nas 10 posições necessárias para o cálculo 
x =  np.ones(10, dtype=int)*x
#Valor de x^n (denominador) em seus 20 termos
y = x**n
#Fatorial de "n" em seus 20 termos (numerador)
fatorial = scipy.special.factorial(n)
#Expressão final
senh = y/fatorial
senh = sum(senh) 
#Resultado
print("O valor do seno hiperbólico é:", senh)
```
